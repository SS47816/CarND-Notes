# Git & Github

## 1. Navigating a Commit History

Clone a reporsitory: `git clone URL`

To check all the commit history: `git log`

with statistics: `git log --stat`

Check differences in two files: `git diff <commit_ID_1> <commit_ID_2>` 

Getting Colored Output: `git config --global color.ui auto`

Revert back to previous version `git checkout commit_ID`

# 2. Creating and Modifying a Repository


## a) Create a Reporsitory

Create a Git repository in a directory that already contains files: `git init`


## b) Commit a Change

View the latest changes: `git status` 

Add a file to the staging area (stage): `git add <file>` 

Remove a file to the staging area (unstage): `git rm --cached <file>` 

Commit: `git commit`

Commit message: `git commit -m "Commit message"` 

Compare the changes between **Working directory** and **Staging area**: `git diff`

Compare the changes between **Staging area** and **Repository**: `git diff --staged`

Discard any changes made in **Working directory** and **Staging area**: `git reset --hard`

Switch to **master** branch `git checkout master`


### Commit Message Style

[Commit Message Style Guide](http://udacity.github.io/git-styleguide/)

```
    type: subject

    body

    footer
    
    
```


#### The Type

**feat:** a new feature

**fix:** a bug fix

**docs:** changes to documentation

**style:** formatting, missing semi colons, etc; no code change

**refactor:** refactoring production code

**test:** adding tests, refactoring test; no production code change

**chore:** updating build tasks, package manager configs, etc; no production code change

## c) Branch

View branches: `git branch`

Create a new branch: `git branch <branch_name>`

Switch to a branch: `git checkout <branch_name>` 

Short cut: `git checkout -b <new_branch_name>`

Visualise the branch structure: `git log --graph --oneline <branch_name_1> <branch_name_2>` 

### Merge Branches

Merge two branches: `git merge <branch_name_1> <branch_name_2>`

Delete a branch: `git branch -d <branch_name>`

Compare the changes with its parent: `git show <commit_ID>.`

#### Merge Conflict

Restore your files to their state before you started the merge by running `git merge --abort`


# 3. Github

## a) Create a Remote Repository

View all the remotes: `git remote` `git remote -v`

Create a remote repo: `git remote add <remote_name> <URL>`

Push: `git push <remote_name> <branch_name>` 

Pull: `git pull <remote_name> <branch_name>` 

Pull = Fetch(`git fetch origin`) + Merge (`git merge master master/origin` )
